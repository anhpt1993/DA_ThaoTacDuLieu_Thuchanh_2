# 0. Import các thư viện cần thiết

```python
import pandas as pd
import numpy as np
```

# 1. Đọc file CSV

```python
df = pd.read_csv("https://s3.ap-southeast-1.amazonaws.com/learn.codegym.vn/media/public/2022/01/20125029/shopeep_koreantop_clothing_shop_data.csv", encoding = "ISO-8859-1")
df.head(10)
```

```python
# Lọc các cột dữ liệu 'date_collected','shop_location','response_time'

df = df.loc[:, ['date_collected','shop_location','response_time']]
df.head(10)
```

```python
df.info()
```

# 2. Tách cột dữ liệu có kiểu chuỗi sử dụng phương thức str.split()

2.1. Tách cột shop_location thành 2 cột "District" và "City"

```python
series_location = df.shop_location.str.split(",")
series_location
```

```python
type(series_location)
```

```python
df["District"] = series_location.str[0].str.strip()
df["City"] = series_location.str[1].str.strip()
df.head(10)
```

# 3. Tách cột dữ liệu có kiểu ngày tháng

3.1. Tách cột date_collected thành 3 cột Day, Month, Year

```python
temp = pd.to_datetime(df.date_collected, format = "%Y-%m-%d")
type(temp)
```

```python
temp
```

```python
df["Day"] = temp.dt.day
df["Month"] = temp.dt.month
df["Year"] = temp.dt.year
df.head(10)
```

# 4. Tách cột dữ liệu có kiểu thời gian

4.1. Tách cột response_time thành 3 cột Hour, Minute, Second

```python
df.info()
```

```python
temp = pd.to_datetime(df.response_time, format = "%H:%M:%S")
type(temp)
```

```python
# Xử lý số liệu sai
df.response_time = df.response_time.str.strip()
```

```python
temp = pd.to_datetime(df.response_time, format = "%H:%M:%S")
type(temp)
```

```python
temp
```

```python
df["Hour"] = temp.dt.hour
df["Minute"] = temp.dt.minute
df["Second"] = temp.dt.second
df.head(10)
```
